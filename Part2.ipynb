```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import requests



url="https://www.amazon.in/dp/B08Z1X5SK4/ref=sspa_dk_detail_2?psc=1&pd_rd_i=B08Z1X5SK4&pd_rd_w=cHYj3&content-id=amzn1.sym.2575ab02-73ff-40ca-8d3a-4fbe87c5a28d&pf_rd_p=2575ab02-73ff-40ca-8d3a-4fbe87c5a28d&pf_rd_r=EYT2KEQCJW2STYTC6S6X&pd_rd_wg=VooM5&pd_rd_r=d594d26e-3eda-4d61-993b-3264aeb05d7a&s=apparel&sp_csd=d2lkZ2V0TmFtZT1zcF9kZXRhaWw"

# Function to scrape product details from the URL
def scrape_product_details(url):
    # Send a GET request to the URL
    # response = requests.get(url)
    status_code=0
    while(status_code !=200):
        response = requests.get(url)
        status_code=response.status_code
        print(response)
        # Create a BeautifulSoup object
    soup = BeautifulSoup(response.text, "html.parser")
    try:
        Description =soup.find("div",attrs={"id":"feature-bullets"}).text
    except:
        Description=""

    # Find the unordered list containing product details
    d={}
    try:
        details_list = soup.find("ul", {"class": "a-unordered-list a-nostyle a-vertical a-spacing-none detail-bullet-list"})
        
            
        # Create an empty dictionary to store the details
        product_details = []

        # Iterate over each list item in the details list
        for item in details_list.find_all("span", {"class":"a-list-item"}):
            # Extract the label and value from each list item
            # label = item.find("span", {"class": "a-text-bold"}).text.strip(": ")
            value = item.text
            
            # Add the label and value to the dictionary
            product_details.append(value)

        for detail in product_details:
            temp=detail.split("\n")
            #print(temp[0].strip(),":" ,temp[-1].strip())
            d[temp[0].strip()]=temp[-1].strip()
        ASIN=d["ASIN"]
        Manufacturer=d["Manufacturer"]
    except:
        ASIN=""
        Manufacturer=""
    try:
        productDescription= soup.find("div",{"id": "productDescription"}).text
    except:
        productDescription=""
    return Description,ASIN,Manufacturer,productDescription,d

# Description, ASIN, Manufacturer, productDescription = scrape_product_details(url)
# print(Description)
# print(ASIN)
# print(Manufacturer)
# print(productDescription)
```
